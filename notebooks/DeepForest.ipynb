Vamos a utilizar el paquete Deepforest https://github.com/weecology/DeepForest para validar su desempeño al identificar árboles a partir de una imagen de prueba

```python
from deepforest import main
from deepforest.visualize import plot_results
import cv2 as cv
import numpy as np

# Carga la clase del modelo y el modelo preentrenado
m = main.deepforest()
m.load_model("weecology/deepforest-tree", revision="main")

# Ruta de la imagen
image_path = '..\data\samples\single_batch_0.tif'
# -- Nota: La imagen debe ser de 3 canales

# Convierte la imagen a un array numpy
image = cv.imread(image_path, cv.IMREAD_UNCHANGED)

# Realiza las predicciones en la imagen
predictions = m.predict_tile(raster_path=image_path,image=image, patch_size=689, patch_overlap=0)

plot_results(predictions, results_color=[66, 100, 245], thickness=6)
```
